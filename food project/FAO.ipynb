```python
for col in foodf.columns:
    pct_missing = foodf[col].isnull().mean()                #to check missing values
    print(f'{col} - {pct_missing :.1%}')
```

```python
foodf = foodf.fillna(foodf.mean())
```

```python
foodf
```

```python
group = foodf.columns[3:56]
foodf[group] = foodf[group].astype(int)
foodf
```

```python
groupfood = foodf.groupby('Area').sum()
```

```python
groupfood
```

```python
cols_to_drop = groupfood.columns[1:52]
df = groupfood.drop(cols_to_drop, axis=1)
```

```python
df
```

```python
df['growth'] = df['Y2013'] - df['Y1961']
```

```python
df
```

```python
negative_values = df[df['growth'] < 0]
print(negative_values)
```

```python
df
```

```python
df['growth %'] = df['Y2013'] / df['Y1961'] * 100
df
```

```python
df['growth %'] = df['growth %'].astype(int)
df
```

```python
df.reset_index(inplace=True)
```
